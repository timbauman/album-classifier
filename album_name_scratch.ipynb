```python
import pandas as pd
# open data file

header_list = ['release_group_id', 'tag_id', 'upvotes', 'updated_at']

release_group_tags = pd.read_csv('/Users/timothybauman/mbdump/release_group_tag', sep='\t', names=header_list)
release_group_tags = release_group_tags[['release_group_id', 'tag_id']]
# release_group_tags.index = release_group_tags.pop('release_group_id')
release_group_tags.head()
```

```python
header_list = ['id', 'mbid', 'name', 'artist_id', 'a', 'b', 'c', 'updated_at']

release_group = pd.read_csv('/Users/timothybauman/mbdump/release_group', sep='\t', names=header_list)
release_group = release_group[['id', 'name', 'artist_id', 'mbid']]
# release_group.index = release_group.pop('id')
release_group.head()
```

```python
header_list = ['id', 'name', 'ref_count']

tags = pd.read_csv('/Users/timothybauman/mbdump/tag', sep='\t', names=header_list)
tags = tags[['id', 'name']]
tags.head()
```

```python
header_list = ['id', 'mbid', 'name', 'a', 'b', 'updated_at']

genre_names = pd.read_csv('/Users/timothybauman/mbdump/genre', sep='\t', names=header_list)

genres = tags[tags['name'].isin(genre_names['name'])]
genres.head()
```

```python
header_list = ['id', 'mbid', 'name', 'sort_name', 'begin', 'b', 'c', 'end', 'd', 'e', 'gender_id', 'area_id', 'h', 'disambiguation', 'j',
               'updated_at', 'k', 'l', 'm']

artists = pd.read_csv('/Users/timothybauman/mbdump/artist', sep='\t', names=header_list, dtype={'disambiguation': str})
artists = artists[['id', 'name', 'mbid']]
# artists.index = artists.pop('id')
artists.head()
```

```python
# format: artist name, album name, tags

albums = release_group.merge(artists, left_on='artist_id', right_on='id', suffixes=('_release_group', '_artist'))
albums.head(50)
```

```python
release_group_genres = release_group_tags.merge(genres, left_on='tag_id', right_on='id', suffixes=('_release_group', '_genre'))
release_group_genres.head()
```

```python
tag_counts = release_group_genres.groupby('name').count().sort_values('release_group_id', ascending=False)
top_tag_counts = tag_counts.head(10)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(top_tag_counts.index)
```

```python
filtered_release_group_genres = release_group_genres[release_group_genres['name'].isin(top_tag_counts.index)]
filtered_release_group_genres.head()
```

```python
release_group_genres_2 = filtered_release_group_genres.groupby(release_group_genres['release_group_id']).aggregate({
    'tag_id': lambda x: set(x),
    'name': lambda x: set(x),
})
release_group_genres_2.head(10)
```

```python
album_genres = albums.merge(release_group_genres_2, left_on='id_release_group', right_on='release_group_id', suffixes=('', '_genre'))
def describe(r): return f"Album name: {r['name_release_group']}\nArtist: {r['name_artist']}"
album_genres['description'] = album_genres.apply(describe, axis=1)
album_genres.head()
```

```python
album_genres_sample = album_genres.sample(100000)
```

```python
def get_y(r): return r['name']
dblock = DataBlock(get_x = describe,get_y = get_y, blocks=(
    TextBlock.from_df('description'),
    MultiCategoryBlock))
# dsets = dblock.datasets(album_genres)
dls = dblock.dataloaders(album_genres_sample)
# dsets.train[0]
dls.one_batch()
```

```python
learner = text_classifier_learner(dls, AWD_LSTM,
# relearn what these arguments mean
    drop_mult=0.3, 
    metrics=partial(accuracy_multi, thresh=0.2)).to_fp16()
```

```python
preds,targs = learner.get_preds()
```

```python
xs = torch.linspace(0.05,0.95,29)
accs = [accuracy_multi (preds, targs, thresh=i, sigmoid=False) for i in xs]
plt.plot(xs,accs);
```

```python
learner.fit(4, lr=2e-3)
```
